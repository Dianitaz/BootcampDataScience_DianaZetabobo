Creación del Dataframe 
Incluye: Datos Duplicados -Valores Nulos -Diferentes Areas y ciudadades

```python
import pandas as pd

# Dataset simulado más completo
data = {
    "nombre": ["Ana", "Luis", "Carlos", "Marta", "Pedro", "Sofia", "Andres", "Laura", "Ana", "Carlos"],
    "edad": [23, 35, 29, 40, None, 31, 28, 45, 23, 29],
    "ciudad": ["Bogotá", "Cali", "Medellín", "Bogotá", "Cali", "Bogotá", "Medellín", "Cali", "Bogotá", "Medellín"],
    "salario": [1750000, 1900000, 2100000, 22000000, 2300000, 2800000, 3250000, 21000000, 4200000, 5100000],
    "area": ["Ventas", "TI", "Ventas", "Gerencia", "TI", "Ventas", "TI", "Gerencia", "Ventas", "Ventas"]
}

df = pd.DataFrame(data)

df
```

Exploración de datos 

head-Como se ven los datos
info -Tipos de datos
describe -Estadisticas
shape- Tamaña de Dataset

```python
df.head()
df.info()
df.describe()
df.shape
```

Limpieza de Datos
Se reemplaza el valor faltante con promedio, para no perder información

```python
df.isnull().sum()
```

```python
df["edad"] = df["edad"].fillna(df["edad"].mean())
```

Verificando la modificación

```python
df.isnull().sum()
```

Eliminar Duplicados
False-No esta repetido
True -Es un duplicado

```python
df.duplicated()
```

Conteo de duplicados donde el esperado es 0, Por lo tanto no requiere limpieza en este paso

```python
df.duplicated().sum()
```

Transformación de Datos 
Se crea una nueva columna  "Salaeio Anual" a partir de salario mensual para un analisis

```python

df["salario_anual"] = df["salario"] * 12

df
```

Clasificación por edad
Transformación de datos numericos en categorias

```python

df["grupo_edad"] = df["edad"].apply(lambda x: "Joven" if x < 30 else "Adulto")

df
```

 Preguntas de análisis

1. ¿Cuál es el salario promedio por ciudad?
2. ¿Qué área tiene el salario promedio más alto?
3. ¿Qué porcentaje de personas gana más que el salario promedio?

```python
# Salario promedio por ciudad
df.groupby("ciudad")["salario"].mean()
```

Varia segun la ciudad, donde se observa que calí presenta el promedio mas alto, seguida de Bogotá y Medellin el promedi mas bajo

## Pregunta 2
¿Cuál es el salario promedio por área?

```python
df.groupby("area")["salario"].mean()
```

Gerencia presenta los salario más altos, mientras que TI y ventas son los mas bajos

## Pregunta 3
¿Qué porcentaje de personas gana más que el salario promedio?

```python
promedio = df["salario"].mean()
promedio
```

```python
mayores_promedio = df[df["salario"] > promedio]

mayores_promedio
```

```python

porcentaje = (len(mayores_promedio) / len(df)) * 100

porcentaje
```
